```python
def split_exp(exp):
    try:
        if len(exp.split('-'))== 2:
            min_exp=exp.split('-')[0]
            max_exp=exp.split('-')[1]
        return pd.Series([min_exp,max_exp])
    except:
        return pd.Series([np.nan,np.nan])
```

```python
df['experience'].apply(split_exp)
```

```python
nm=pd.DataFrame(df['experience'].str.contains('Not Mentioned'))
```
